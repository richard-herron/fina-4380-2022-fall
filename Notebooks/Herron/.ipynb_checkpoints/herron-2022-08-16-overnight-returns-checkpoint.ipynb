The overnight effect refers to the fact that, over at least the past three decades, investors have earned 100% or more of the return on a wide range of risky assets when the markets are closed, and, as sure as day follows night, have earned zero or negative returns for bearing the risk of owning those assets during the daytime, when markets are open. # How to separate daily returns into intraday and overnight returns

Yahoo! Finance provides easy acces to high-quality open, high, low, close (OHLC) and adjusted close price data.
However, Yahoo! Finance does not provide overnight or instraday returns directly.
Therefore, we need to use math to decompose daily returns into overnight and intraday returns.

Daily returns are defined as (adjusted) closing price to (adjusted) closing price returns.
Therefore, daily returns consist of overnight returns compounded with the intraday returns from the next day $$(1 + R_{daily}) = (1 + R_{overnight}) \cdot (1 + R_{intraday}),$$ which we can rearrange to calculate overnight returns as $$\frac{1 + R_{daily}}{1 + R_{intraday}} - 1 = R_{overnight}.$$

We can calculate daily and intraday returns from Yahoo! Finance data as $$R_{daily} = \frac{Adj\ Close_{t} - Adj\ Close_{t-1}}{Adj\ Close_{t-1}}$$ and $$R_{intraday} = \frac{Close - Open}{Open}.$$

# Compare S&P 500 ETF intraday and overnight returns

We will explore the overnight returns anomaly first with the S&P 500 ETF (ticker SPY).
***Note, many of the chained methods below are very long.
We can insert line breaks inside a chained method if we wrap the whole chain in parentheses.***

```python
spy = yf.download(tickers='SPY', session=session)
```

We use the formulas above to calculate daily, intraday, and overnight returns.
I will name these returns `Rd`, `Ri`, and `Ro` for brevity.

```python
spy['Rd'] = spy['Adj Close'].pct_change()
spy.eval('Ri = (Close - Open) / Open', inplace=True)
spy.eval('Ro = (1 + Rd) / (1 + Ri) - 1', inplace=True)
```

We can calculate intraday returns for the first day (1993-01-29).
However, we cannot calculate daily and overnight returns for this first day because we do have the previous adjusted close.
Therefore, we will drop the first day to make sure we use properly align returns.

```python
spy_returns = spy[['Ro', 'Ri']].dropna()
```

Now we can compound intraday and overnight returns separately.

```python
labels = {
	'Rd': 'Daily Returns',
	'Ri': 'Intraday Returns',
	'Ro': 'Overnight Returns'
}
```

```python
spy_returns.add(1).cumprod().sub(1).mul(100).rename(columns=labels).plot()
plt.ylabel('Cumulative Return (%)')
plt.title('Overnight Versus Intraday Returns for the S&P 500 ETF (SPY)')
plt.show()
```

There is a stark difference between the overnight and intrday returns!
The total returns in percent are:

```python
(
	spy_returns
	.add(1)
	.prod()
	.sub(1)
	.mul(100)
	.rename(labels)
	.to_frame('Total Return (%)')
	.pipe(display_table)
)
```

One explanation for this stark difference is that risk also occurs overnight (e.g., earnings announcements, macroeconomic announcements, etc.).
However, on this stark difference persists on risk-adjusted basis (i.e., Sharpe Ratios):

```python
(
	spy_returns
	.mean()
	.div(
		spy_returns
		.std()
	)
	.mul(np.sqrt(252))
	.rename(labels)
	.to_frame('Sharpe Ratio')
	.pipe(display_table)
)
```

The jury is still out on what causes this difference.
However, this difference likely persists because transaction costs and short-term capital gains taxes make it costly to separate overnight and intraday returns.

# Compare single stocks intraday and overnight returns

Does the large difference between overnight and intraday returns exist for single stocks, too?
We can repeat the analysis above for the 500 constituents of the S&P 500.

```python
wiki = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
```

Wikipedia uses `.` to separate tickers and suffixes but Yahoo! Finance expects `-`.
We can use the `.replace()` method to replace `.` with `-` and the `.to_list()` to create the list of tickers that `yf.download()` expects.

```python
tickers = wiki[0]['Symbol'].replace(to_replace='\.', value='-', regex=True).to_list()
```

Note, we are pushing the limits of Yahoo! Finance data, so we limit our single-stock analysis to the last five years.

```python
sp500 = yf.download(tickers, session=session, period='5y')
```

It will easiest to caculates daily, intraday, and overnight returns in separate data frames, then combine.
This approach has more compact code and makes it easier to drop days without all all three returns.

```python
Rd = sp500['Adj Close'].pct_change()
Ri = (sp500['Close'] - sp500['Open']) / sp500['Open']
Ro = (1 + Rd) / (1 + Ri) - 1
```

```python
sp500_returns = pd.concat(objs=[Ro, Ri], keys=['Ro', 'Ri'], names=['Variable', 'Date'])
sp500_returns.columns.name = 'Ticker'
```

```python
sp500_returns = sp500_returns.T.stack().dropna()
```

We see below that the mean five-year overnight return is 105% versus a mean five-year intraday return of only 16%.

```python
(
	sp500_returns
	.add(1)
	.groupby('Ticker')
	.prod()
	.sub(1)
	.mul(100)
	.describe()
	.T
	.rename(index=labels)
	.pipe(display_table)
)
```

As well, the risk-return tradeoff favors overnight returns.

```python
(
	sp500_returns
	.mean()
	.div(
		sp500_returns
		.std()
	)
	.mul(np.sqrt(252))
	.rename(labels)
	.to_frame('Share Ratio')
	.pipe(display_table)
)
```

The plot above would be difficult to interpret for 500 S&P 500 constituents and computationally expensive.
Instead, we can calculate summary statistics for the relative value of $1 invested in overnight versus intraday returns.
We take the log of the relative value to the positive values indicate overnight > intraday, negative values indicate overnight < intraday, and zeros indicate overnight = intraday.

```python
_ = sp500_returns.add(1).groupby('Ticker').prod()
relative_value = np.log(_['Ro'] / _['Ri'])
(
	relative_value
	.to_frame('log(Overnight Value/Intraday Value)')
	.describe()
	.pipe(display_table)
)
```

More than 75% of stocks have overnight returns that exceed their intraday returns (i.e., at the 25%, the log of the ratio of overnight value to intraday value is 0.06).
We can visualize this log ratio as a [swarmplot](https://seaborn.pydata.org/generated/seaborn.swarmplot.html).

```python
import seaborn as sns
```

```python
sns.swarmplot(relative_value)
plt.xlabel('log(Overnight Value/Intraday Value)')
plt.axvline(0, color='red')
plt.title('Distribution of the log(Overnight Value/Intraday Value)')
plt.show()
```
