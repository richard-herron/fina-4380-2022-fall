```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.rcParams['figure.dpi'] = 150
np.set_printoptions(precision=2, suppress=True)
pd.options.display.float_format = '{:,.2f}'.format
```

```python
# this function pretty prints data frames in the PDF export
# e.g., display_table(df) pretty prints df in the PDF export
import warnings
def display_table(x):
    with warnings.catch_warnings() and pd.option_context('display.latex.repr', True):
        warnings.simplefilter(action='ignore', category=FutureWarning)
        display(x)
```

```python
import yfinance as yf
import pandas_datareader as pdr
import requests_cache
session = requests_cache.CachedSession()
```

```python
ff0 = pdr.get_data_famafrench('F-F_Research_Data_Factors', start='1900-01-01', session=session)
```

```python
ff = ff0[1]
```

```python
ff /= 100
```

```python
mkt = ff['Mkt-RF'] + ff['RF']
```

```python
mkt.index = mkt.index.year
```

```python
df = pd.DataFrame(
    data=np.zeros((len(mkt.index), len(mkt.index))), 
    index=mkt.index.copy(), 
    columns=mkt.index.copy()
)
df[:] = np.nan
```

```python
df.index.name = 'From Date'
df.columns.name = 'To Date'
```

```python
def annret(r):
    return r.add(1).prod()**(1 / r.shape[0]) - 1
```

```python
for r in df.index:
    for c in df.index[df.index >= r]:
        df.loc[r, c] = annret(mkt.loc[r:c])
```

```python
s = pd.Series(np.diag(df.iloc[:-19, 19:]), index=[df.index[:-19], df.columns[19:]])
```

```python
import seaborn as sns
```

```python
ax = sns.heatmap(df.mul(100), square=True, robust=True, cmap='RdYlGn', cbar_kws={'label': 'Annualized Return (%)'})

for j in df.columns[19:]:
    j -= df.columns[0]
    i = j - 19
    rect = plt.Rectangle([j, i], 1, 1, color="black", linewidth=0.5, fill=False, clip_on=False)
    ax.add_patch(rect)

xy_max = list((np.array(s.idxmax()) - 1927)[::-1])
rect_max = plt.Rectangle(xy_max, 1, 1, color="black", linewidth=1, fill=False, clip_on=False)
ax.add_patch(rect_max)

ax.text(xy_max[0]+0.5, xy_max[1]-0.5, "Best return: {:.1f}%".format(s.mul(100).max()), **{'fontsize': 'xx-small'})

xy_min = list((np.array(s.idxmin()) - 1927)[::-1])
rect_min = plt.Rectangle(xy_min, 1, 1, color="black", linewidth=1, fill=False, clip_on=False)
ax.add_patch(rect_min)

ax.text(xy_min[0]+0.5, xy_min[1]-0.5, "Worst return: {:.1f}%".format(s.mul(100).min()), **{'fontsize': 'xx-small'})

plt.title('It is when you start and finish')

plt.show()
```
