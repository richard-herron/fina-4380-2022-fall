# Write a function to calculate annualized total returns (CAGRs)

From [Investopedia](https://www.investopedia.com/terms/c/cagr.asp):

>  What Is the Compound Annual Growth Rate (CAGR)?
> 
> The compound annual growth rate (CAGR) is the rate of return (RoR) that would be required for an investment to grow from its beginning balance to its ending balance, assuming the profits were reinvested at the end of each period of the investmentâ€™s life span. 

The CAGR or annualized total return formula is $$CAGR = \left[ \prod_{t=1}^{T} (1 + r_t) \right]^{\frac{1}{T}} - 1.$$

We will write a function called `annret()` to calculate CAGR or annualized total return.

```python
def annret(r):
    return r.add(1).prod()**(1/r.shape[0]) - 1
```

# Populate a data frame with the annualized total returns

```python
df = pd.DataFrame(
    data=np.zeros((mkt.shape[0], mkt.shape[0])),
    index=mkt.index.copy(),
    columns=mkt.index.copy()
)
```

Matplotlib and seaborn will use index and column names as axis labels.
***I changed these labels after class to make the axis labels match my snappy plot titles.***

```python
df.index.name = 'Enter Year'
df.columns.name = 'Exit Year'
```

```python
df[:] = np.nan
```

```python
for r in df.index:
    for c in df.columns[df.columns >= r]:
        df.loc[r, c] = annret(mkt.loc[r:c])
```

# Write a function to plot the annotated heatmap

```python
import seaborn as sns
```

```python
def my_heatmap(x, title='', hp=20):

	# basic heat map
	ax = sns.heatmap(x.mul(100), robust=True, cmap='RdYlGn', cbar_kws={'label': 'Annualized Return (%)'})
	plt.title(title)

	# select diagonal values that correspond to holding period
	hp -= 1
	diag = pd.Series(np.diag(x.iloc[:-1*hp, hp:]), index=[x.index[:-1*hp], x.columns[hp:]])

	# loop over diagonal
	for (i, j), r in diag.iteritems():

		# axes are relative to first years
		i -= x.index[0]
		j -= x.columns[0]

		# select colors and annotate outliers
		if r == diag.min():
			color = 'blue'
			annot = 'Lowest: {:.1f}%'.format(100*r)
			ax.text(j+1, i, annot, **{'fontsize': 'x-small', 'color': color})
		elif r == diag.max():
			color = 'blue'
			annot = 'Highest: {:.1f}%'.format(100*r)
			ax.text(j+1, i, annot, **{'fontsize': 'x-small', 'color': color})
		else:
			color = 'black'

		rect = plt.Rectangle([j, i], 1, 1, color=color, linewidth=1, fill=False, clip_on=False)
		ax.add_patch(rect)
```

```python
my_heatmap(df, title='It Matters When You Enter And Exit The Market\n(Nominal Returnsor Not-Inflation-Adjusted Returns)')
```

# Inflation adjust market returns and re-use our heatmap function

***I tweaked the following two cells after class.***
These tweaks:

1. Use an inflation measure with a longer time series (`CPIAUCNS`)
1. Correctly align the inflation values to each year (I forgot `.shift(freq='-1d')` in class)

```python
cpi = pdr.get_data_fred('CPIAUCNS', start='1900-01-01', session=session)
```

```python
infl = cpi.shift(freq='-1d').resample(rule='A', kind='period').last().pct_change()['CPIAUCNS']
```

```python
infl.index = infl.index.year
```

```python
mkt_ia = (1 + mkt) / (1 + infl) - 1
```

```python
mkt_ia.dropna(inplace=True)
```

```python
df_ia = pd.DataFrame(
    data=np.zeros((mkt_ia.shape[0], mkt_ia.shape[0])),
    index=mkt_ia.index.copy(),
    columns=mkt_ia.index.copy()
)
```

***I changed these labels after class to make the axis labels match my snappy plot titles.***

```python
df_ia.index.name = 'Enter Year'
df_ia.columns.name = 'Exit Year'
```

```python
df_ia[:] = np.nan
```

```python
for r in df_ia.index:
    for c in df_ia.columns[df_ia.columns >= r]:
        df_ia.loc[r, c] = annret(mkt_ia.loc[r:c])
```

```python
my_heatmap(df_ia, title='It Matters When You Enter And Exit The Market\n(Real Returns or Inflation-Adjusted Returns)')
```
