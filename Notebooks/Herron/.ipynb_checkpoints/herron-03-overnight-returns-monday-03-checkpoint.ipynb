```python
import yfinance as yf
import pandas as pd
import requests_cache
import matplotlib.pyplot as plt
import numpy as np
session = requests_cache.CachedSession()
```

```python
plt.rcParams['figure.dpi'] = 150
np.set_printoptions(precision=4, suppress=True)
pd.options.display.float_format = '{:.4f}'.format
```

```python
df = yf.download(tickers='SPY', session=session)
```

1. Calculate intraday returns as `Open` to `Close` returns
2. Calculate total returns as `Adj Close` to `Adj Close` returns
3. The overnight return is the portion of the total return not explained by intraday returns

```python
df['R_intraday'] = df['Close'] / df['Open'] - 1
df.eval('R_intraday_alt = Close / Open - 1', inplace=True)
```

```python
np.allclose(df['R_intraday'], df['R_intraday_alt'])
```

```python
df['R_total'] = df['Adj Close'].pct_change()
```

Overnight returns are the portion of total returns that are not due to intraday returns.

$$1 + R_{total} = (1 + R_{intraday}) \times (1 + R_{overnight}) \implies  R_{overnight} = \frac{1 + R_{total}}{1 + R_{intraday}} - 1$$

```python
df['R_overnight'] = (1 + df['R_total']) / (1 + df['R_intraday']) - 1
```

```python
labels = {
    'R_intraday': 'Intraday Returns', 
    'R_overnight': 'Overnight Returns',
    'R_total': 'Total Returns'
}
```

```python
df[['R_intraday', 'R_overnight']].rename(columns=labels).mean().mul(252 * 100).plot(kind='bar')
plt.xticks(rotation=0)
plt.ylabel('Annualized Mean Daily Return (%)')
plt.title('Comparison of Intraday and Overnight Returns\nBased on the SPY ETF')
plt.show()
```

```python
df[['R_intraday', 'R_overnight']].dropna()
```

***I added the `_` step after class to simplify the code.***

```python
_ = df[['R_intraday', 'R_overnight']].dropna()
_.rename(columns=labels).add(1).cumprod().plot()
plt.ylabel('Value ($)')
plt.title(
    'Comparison of Intraday and Overnight Returns' +
    '\n' +
    'Value of $1 Invested in SPY on ' + 
    _.index[0].strftime('%b %d, %Y')
)
plt.show()
```

Full-sample returns from intraday and overnight returns in percent:

```python
df[['R_intraday', 'R_overnight']].dropna().add(1).prod().sub(1).mul(100)
```

What if we repeat this exercise for every stock in the S&P 500 instead of for the S&P 500 ETF?
The S&P500 ETF is value-weighted, so the results above could be due to a handful of large stocks.
Note that this solutions relies on a few tools from chapters 8 and 10 of McKinney.

First, we need tickers for the S&P 500 components.
Note that `pd.read_html()` makes reading Wikipedia *easy*!

```python
components = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
```

```python
tickers = components['Symbol']
```

```python
tickers
```

To simplify our analysis, use only the last one year of data via `period = '1y'`.

```python
df2 = yf.download(tickers=tickers.to_list(), period='1y', session=session)
```

```python
df2.columns.names = ['Variable', 'Ticker']
```

```python
df2_long = df2.stack().swaplevel().sort_index(level=['Ticker', 'Date'])
```

```python
df2_long['R_intraday'] = df2_long['Close'] / df2_long['Open'] - 1
```

```python
df2_long['R_total'] = df2_long.groupby('Ticker')['Adj Close'].pct_change()
```

```python
df2_long['R_overnight'] = (1 + df2_long['R_total']) / (1 + df2_long['R_intraday']) - 1
```

We have some choice in how to aggregate these data.
First, we can find the mean return for each ticker:

```python
df2_long[['R_intraday', 'R_overnight']].dropna().rename(columns=labels).groupby('Ticker').mean()
```

Then, we can take the mean across these tickers by chaining another `.mean()` method.

```python
df2_long[['R_intraday', 'R_overnight']].dropna().rename(columns=labels).groupby('Ticker').mean().mean()
```

```python
_ = df2_long[['R_intraday', 'R_overnight']].dropna()
_ .rename(columns=labels).groupby('Ticker').mean().mean().mul(252 * 100).plot(kind='bar')
plt.xticks(rotation=0)
plt.ylabel('Annualized Mean Daily Return (%)')
plt.title('Comparison of Intraday and Overnight Returns')
plt.show()
```

We could also create an *equal-weighted* portfolio of the S&P 500 components, then calculate cumulative return of this portfolio.

```python
_ = df2_long[['R_intraday', 'R_overnight']].dropna()
_.unstack(0).groupby('Variable', axis=1).mean().rename(columns=labels).add(1).cumprod().plot()
plt.ylabel('Value ($)')
plt.title(
    'Comparison of Intraday and Overnight Returns' +
    '\n' +
    '$1 Invested in Equal-Weighted S&P 500 on ' + _.index.get_level_values(1)[0].strftime('%b %d, %Y')
)
plt.show()
```
