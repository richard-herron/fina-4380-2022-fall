***Question:***
How do I calculate beta from Yahoo! Finance data?

***Answer:***
Here is a simple walkthrough from class.
This walkthrough uses some approximations that we will improve on later in the semester:

1. Use the S&P 500 ETF (SPY) to approximates the market return
1. Ignores the risk-free rate, and uses raw returns to estimate beta ($\beta$) instead of _excess_ returns
1. Estimates beta as the ratio of covariances and variances, instead of estimating a linear regression (i.e., $\beta_{TSLA} = \frac{Cov(R_{TSLA}, R_{SPY})}{Var(R_{SPY})}$); later we will estimate linear regression to estimate the intercept (alpha or $\alpha$)

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import requests_cache
session = requests_cache.CachedSession()
```

```python
plt.rcParams['figure.dpi'] = 150
pd.options.display.float_format = '{:.4f}'.format
```

```python
df = yf.download(tickers='TSLA SPY', session=session)
```

```python
returns = df['Adj Close'].pct_change().dropna()
```

```python
returns.plot(x='SPY', y='TSLA', kind='scatter', marker='.', alpha=0.25, grid=True)
plt.xlabel('Daily Return on SPY')
plt.ylabel('Daily Return on TSLA')
plt.title('Beta Plot for TSLA')
plt.show()
```

Systematic risk (beta or $\beta$) is this slope of the best-fit line through the scatter plot above: $$\beta_{TSLA} = \frac{Cov(R_{TSLA}, R_{SPY})}{Var(R_{SPY})}$$

```python
beta = returns.cov().loc['TSLA', 'SPY'] / returns.var().loc['SPY']
print('TSLA beta: {0:.2f}'.format(beta))
```

We can use the seaborn plotting package to easily add the best-fit line to the beta plot:
(***Note:*** After class I looked up how to add the grid to more closely match the matplotlib plot.)

```python
import seaborn as sns
sns.set_style("whitegrid")
sns.regplot(x='SPY', y='TSLA', data=returns, marker='.', scatter_kws={'alpha':0.25})
plt.xlabel('Daily Return on SPY')
plt.ylabel('Daily Return on TSLA')
plt.title('Beta Plot for TSLA')
plt.show()
```
